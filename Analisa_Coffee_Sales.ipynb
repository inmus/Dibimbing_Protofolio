Observasi:

Secara keseluruhan, nilai minimum dan maksimum masuk akal untuk setiap kolom.

Mean ~ 50% (Median) di kolom hour_of_day, money,	Weekdaysort,	Monthsort, menunjukkan distribusi normal.

```python
# Mengatasi missing value
for column in data.columns:
    if data[column].dtype == 'object':
        # Jika kolom bertipe object, isi dengan mode
         data[column].fillna( data[column].mode()[0], inplace=True)
    else:
        # Jika kolom bertipe numerik, isi dengan mean
         data[column].fillna( data[column].mean(), inplace=True)
```

```python
data.info()
```

```python
# cek kembali missing value
data.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weekday_sales = data.groupby('Weekday')['money'].sum().reindex(
    ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(x=weekday_sales.index, y=weekday_sales.values, palette='viridis')
plt.title('Total Penjualan per Hari dalam Seminggu')
plt.xlabel('Hari')
plt.ylabel('Total Penjualan (R)')
plt.tight_layout()
```
